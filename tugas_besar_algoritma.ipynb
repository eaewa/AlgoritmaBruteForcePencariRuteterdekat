```python
import ipywidgets as widgets
from IPython.display import display

jarak = {
    ('a', 'b'): 2.9, ('b', 'a'): 2.9, ('a', 'c'): 3.6, ('c', 'a'): 3.6, ('a', 'd'): 3.0, ('d', 'a'): 3.0, ('a', 'e'): 5.0, ('e', 'a'): 5.0, ('b', 'c'): 1.8, ('c', 'b'): 1.8, ('b', 'd'): 1.0, ('d', 'b'): 1.0, ('b', 'e'): 1.5, ('e', 'b'): 1.5,
    ('c', 'd'): 1.6, ('d', 'c'): 1.6, ('c', 'e'): 6.6, ('e', 'c'): 6.6, ('d', 'e'): 7.0, ('e', 'd'): 7.0
}

titik = ['a', 'b', 'c', 'd', 'e', 'a']

dropdowns = []
for i in range(len(titik)):
    dropdown = widgets.Dropdown(
        options=titik,
        description=f"Tujuan ke-{i + 1}:",
        style={'description_width': 'initial'}
    )
    dropdowns.append(dropdown)

button = widgets.Button(description="Hitung")

output = widgets.Output()

def hitung_jarak_termurah(b):
    rute = [dropdown.value for dropdown in dropdowns]

    jarak_terpendek = float('inf')
    rute_terpendek = None

    for r in [rute]:
        jarak_rute = 0
        for i in range(len(r) - 1):
            titik1 = r[i]
            titik2 = r[i + 1]
            jarak_rute += jarak[(titik1, titik2)]
        if jarak_rute < jarak_terpendek:
            jarak_terpendek = jarak_rute
            rute_terpendek = r

    biaya = (jarak_terpendek / 10) * 10000

    with output:
        output.clear_output()
        print("Jarak terpendek adalah:", jarak_terpendek, "km")
        print("Rute terpendek:", ' -> '.join(rute_terpendek))
        print("Biaya:", biaya)

button.on_click(hitung_jarak_termurah)

widgets.VBox(dropdowns + [button, output])
```
